```python
import numpy as np
import matplotlib.pyplot as plt
x_pos=0
y_pos=0
a_pos=0
b_pos=0
x_direct=1
y_direct=1
a_direct=-1
b_direct=1
fig,ax=plt.subplots(figsize=(50,20))
ax.quiver(x_pos,y_pos,x_direct,y_direct,color='y')
ax.quiver(a_pos,b_pos,a_direct,b_direct)
ax.set_title('quiver plot with one arrow')
plt.show()
```

```python
x_pos=0
y_pos=0
x_direct=1
y_direct=1
fig,ax=plt.subplots(figsize=(5,5))
ax.quiver(x_pos,y_pos,x_direct,y_direct,color='b',scale=5)
ax.set_title('quiver plot with one arrow')
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize =(14, 8))
x = np.arange(0, 2.2, 0.2)
y = np.arange(0, 2.2, 0.2)
X, Y = np.meshgrid(x, y)
u = np.cos(X)*Y
v = np.sin(y)*Y
n = -2
color = np.sqrt(((v-n)/2)*2 + ((u-n)/2)*2)
ax1.quiver(X, Y, u, v, color, alpha = 0.8)
ax1.xaxis.set_ticks([])
ax1.yaxis.set_ticks([])
ax1.axis([-0.2, 2.3, -0.2, 2.3])
ax1.set_aspect('equal')
ax1.set_title('meshgrid function')
```
