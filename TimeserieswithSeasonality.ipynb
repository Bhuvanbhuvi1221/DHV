```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01"]),
    "Value":[10,30,15,50,15]
}
df = pd.DataFrame(data)
df.set_index("Date",inplace=True)

plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='o')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Seasonality")
plt.grid(True)
plt.show()
```
