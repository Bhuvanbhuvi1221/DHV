```python
tracks['duration'] = tracks['duration_ms'].apply (lambda x : round(x/1000))
tracks.drop('duration_ms', inplace = True, axis=1)
tracks.duration.head()
```

```python
cm = tracks.drop(['time_signature','mode','track_id','artists','album_name','track_genre'], axis=1).corr(method = 'pearson')
plt.figure(figsize=(14,6))
cmap = sns.heatmap(cm, annot = True, fmt = '.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
cmap.set_title('Correlation Heatmap between Variable')
cmap.set_xticklabels(cmap.get_xticklabels(), rotation=90)
```

```python
sam = tracks.sample(int(0.004 * len(tracks)))
len(sam)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sam, y='loudness', x='acousticness', color='c').set(title='Loudness vs Acousticness Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sam, y='popularity', x='acousticness', color='b').set(title='Popularity vs Acousticness Correlation')
```
